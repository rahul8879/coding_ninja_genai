```python
from email_classifier_use_case.utils import process_email
```

```python
email = {'id': 2,
 'subject': 'App crashes on settings page',
 'sender': 'user@bigclient.com',
 'body': 'Every time I open Settings the app crashes. Chrome 120, Windows 11.',
 'correct': 'Technical',
 'model_output': 'Technical'}
```

```python
data = process_email(email)
```

```python
import json
row = {
    'email_id': data['email_id'],
    'category': data['classification']['category'],
    'urgency': data['classification']['urgency'],
    'confidence': data['classification']['confidence'],
    'routed_to': data['routed_to'],
    'drafted_response': data['drafted_response'],
    'audit_trail': json.dumps(data['audit_trail']),
}

import pandas as pd
df = pd.DataFrame([row])
df.to_csv('processed_emails.csv', index=False)
```
